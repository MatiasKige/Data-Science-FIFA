Se realiza el preprocesamiento de los datos acordes a valores nulos.

```python
def fill_na_in_column(df, column, fill_value):                                  # Convertir de valor nulo a algo.
    df[column].fillna(fill_value, inplace=True)
    return df

def drop_na_in_column(df, column):                                              # Eliminar registros con valores nulos.
    return df.dropna(subset=[column])

def map_column_values(df, column, mapping_dict):                                # Reemplazar valores.
    df[column] = df[column].apply(lambda value: mapping_dict.get(value, value))
    return df

def convert_to_datetime(df, columns):                                            # Convertir a tipo datetime.
    for column in columns:
        df[column] = pd.to_datetime(df[column])
    return df

def convert_value(value):                                                       # Convertir valor int a float o float a int.
    if isinstance(value, float):
        return int(value)
    elif isinstance(value, int):
        return float(value)
    else:
        return value
def convert_types(df, column):
      df[column] = df[column].apply(convert_value)
      return df

def preprocess_data(df):                                                         # Preprocesamiento de los datos.
    gender_mapping = {
        "Right": "R",
        "Left": "L"}
    return (
        df.pipe(fill_na_in_column, "nation_position", "No Convocado")
        .pipe(fill_na_in_column, "nation_jersey_number", "No Convocado")
        .pipe(fill_na_in_column, "club_loaned_from", "No prestado")
        .pipe(fill_na_in_column, "goalkeeping_speed", "-")
        .pipe(fill_na_in_column, "pace", "-")
        .pipe(fill_na_in_column, "shooting", "-")
        .pipe(fill_na_in_column, "defending", "-")
        .pipe(fill_na_in_column, "dribbling", "-")
        .pipe(fill_na_in_column, "passing", "-")
        .pipe(fill_na_in_column, "physic", "-")
        .pipe(drop_na_in_column, "release_clause_eur")
        .pipe(drop_na_in_column, "club_jersey_number")
        .pipe(map_column_values, "preferred_foot", gender_mapping)
        .pipe(convert_types, "fifa_version")                                    # float a int
        .pipe(convert_types, "nation_jersey_number")                            #
        .pipe(convert_types, "club_jersey_number")                              #
        .pipe(convert_types, "club_contract_valid_until_year")                  #
        .pipe(convert_types, "mentality_composure")                             #
        .pipe(convert_types, "shooting")                                        #
        .pipe(convert_types, "goalkeeping_speed")                               #
        .pipe(convert_types, "weight_kg")                                       # int a float
        .pipe(convert_to_datetime, ["dob", "club_joined_date"])                 # datetime
    )

df = preprocess_data(df)
df[["short_name","nation_position","nation_jersey_number", "club_loaned_from","goalkeeping_speed","release_clause_eur","dob"]]
```
