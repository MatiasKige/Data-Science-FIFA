```python
df = pd.read_csv("/content/drive/MyDrive/Curso Data Science/Desafios entregables/Proyecto de prueba FIFA24/male_players.csv")

# Eliminacion de columnas irrelevantes
df = df.drop(columns=['player_id','player_url',"value_eur","fifa_update","update_as_of","club_team_id","league_id","league_level","club_position","nationality_id","nation_team_id","player_tags","real_face","player_traits"])

# Creacion de informacion monetaria relevante para el club.

print(df.shape)
df.head()
```

## Valores nulos

Se realiza una busqueda de valores ausentes (NaN) en las columnas. Se realizara un analisis desde la columna con mas valores continuando con la segunda con mayor cantidad de valores.

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

La primera aparicion de valores nulos es para la columna de "nation_jersey_number" (numero de camiseta en su seleccion nacional) misma cantidad de nulos que "nation_posicion" (posicion del jugador en su seleccion nacional).

```python
df_sin_posicion_seleccion = df[df["nation_position"].isnull()]
print(df_sin_posicion_seleccion["nation_position"].count())
print(df_sin_posicion_seleccion["nation_jersey_number"].count())
```

Al comparar ambos registros notamos que no hay informacion de ellos en sus selecciones, esto significa que no pertenecen al grupo de convocados por su seleccion nacional. Procedemos a modificar estos valores NaN por "No convocado".

```python
df["nation_position"] = df["nation_position"].fillna("No convocado")
df["nation_jersey_number"] = df["nation_jersey_number"].fillna("No convocado")
df[["short_name","nation_position","nation_jersey_number"]].head()
```

La siguiente columna con mayor cantidad de valores nulos es "club_loaned_from" (Club que presta al jugador). Aqui yacen los clubes que dan a prestamo algun jugador, es de obviar que habra muchisimos faltantes ya que no todos los jugadores son prestamos. Procedemos a modificar el valor Nan por "No prestado".

```python
df["club_loaned_from"] = df["club_loaned_from"].fillna("No prestado")
df[["short_name","nation_position","club_loaned_from"]].head()
```

La columna "goalkeeping_speed" (velocidad de arquero) es un valor unico para jugadores de posicion GK (arquero).

```python
df_arqueros = df[df["player_positions"] == "GK"]
print ("Cantidad de arqueros: ",df_arqueros["player_positions"].count())
print ("Cantidad total jugadores: ",df["player_positions"].count())
```

Si restamos la cantidad de arqueros a la cantidad total de jugadores nos devuelve exactamente 159.997 jugadores, justamente la cantidad de valores nulos que aparecen en la columna "goalkeeping_speed". Procedemos a modificar los valores Nan por "-".

```python
df["goalkeeping_speed"] = df["goalkeeping_speed"].fillna("-")
df[["short_name","nation_position","goalkeeping_speed"]].head()
```

Notamos varios columnas con el mismo numero de valores faltantes, veremos si estos coinciden los registros.

```python
df_sin_pace = df[df["pace"].isnull()]
df_sin_pace[["short_name","pace","shooting","defending","dribbling","passing","physic","player_positions"]].head()
```

Detectamos gracias a éste agrupamiento que los jugadores que no poseen estas caracteristicas "pace","shooting","defending", etc, se debe a que no son valores propios de ARQUEROS, si nos fijamos bien, 20024 son la cantidad de valores nulos de estas columnas, siendo 20024 tambien la cantidad total de arqueros en los registros. Procedemos a modificar los valores NaN por "-".

```python
df["pace"] = df["pace"].fillna("-")
df["shooting"] = df["shooting"].fillna("-")
df["defending"] = df["defending"].fillna("-")
df["dribbling"] = df["dribbling"].fillna("-")
df["passing"] = df["passing"].fillna("-")
df["physic"] = df["physic"].fillna("-")
df[["short_name","nation_position","goalkeeping_speed","pace","shooting","defending","dribbling","passing","physic","player_positions"]].head()
```

"release_clause_eur" (valor de clausula de liberacion)

```python
print ("Valores reales:",df["release_clause_eur"].count())
print ("Valores nulos: ",df["release_clause_eur"].isnull().sum())

df_sin_clausula = df[df["release_clause_eur"].isnull()]
df_sin_clausula[["short_name","release_clause_eur","wage_eur","club_contract_valid_until_year","overall"]].head()
```

No pudimos encontrar a que se deben los valores faltantes en la columna "release_clause_eur", aun conservando los valores "wage_eur"(salario). De querer proceder con la negociacion con uno de estos jugadores puede haber inconvenientes por la falta de informacion. Procedemos a eliminar estos registros de nuestro dataframe.

```python
print ("Antes:",df["release_clause_eur"].isnull().sum())
df = df.dropna(subset=["release_clause_eur"])
print ("Ahora:",df["release_clause_eur"].isnull().sum())
df = df.dropna(subset=["club_jersey_number"])
```

Esta eliminacion de registros se ha llevado consigo el resto de valores nulos apreciables del dataframe, dandonos esto mas razon para no contar con ellos.

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

Ya no hay valores nulos en nuestra informacion.

## Conversion de tipos

```python
df['club_jersey_number'] = df["club_jersey_number"].astype(int)
df["weight_kg"] = df["weight_kg"].astype(float)
df["club_contract_valid_until_year"] = df["club_contract_valid_until_year"].astype(int)
df['nation_jersey_number'] = df['nation_jersey_number'].apply(lambda x: int(x) if isinstance(x, float) else x)
df["pace"] = df["pace"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["shooting"] = df["shooting"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["defending"] = df["defending"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["dribbling"] = df["dribbling"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["passing"] = df["passing"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["physic"] = df["physic"].apply(lambda x: int(x) if isinstance(x,float) else x)
df["mentality_composure"] = df["mentality_composure"].astype(int)
df["goalkeeping_speed"] = df["goalkeeping_speed"].apply(lambda x: int(x) if isinstance(x,float) else x)
df['preferred_foot'] = df['preferred_foot'].replace({'Right': 'R', 'Left': 'L'})
df['dob'] = pd.to_datetime(df['dob'])
df["club_joined_date"] = pd.to_datetime(df["club_joined_date"])
df['fifa_version'] = df["fifa_version"].astype(int)
```

Las predicciones a continuacion se realizan en valor numerico, por lo cual filtramos en nuestro DataFrame aquellas columnas que contengan valores numericos en sus filas.

```python
df_num = df_24.select_dtypes(include='number')
df_num.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

Y = df_num['overall']
X = df_num.drop('overall', axis=1)    # el eje 1 se refiere a las columnas, mientras que el eje 0 se refiere a las filas.
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=5)

model = LinearRegression()

sfs = SFS(estimator=model,
          k_features=3,  # Número de columnas a seleccionar
          forward=True,
          floating=False,
          scoring='neg_mean_squared_error',  # Error cuadrático medio negativo
          cv=5)     # Número de divisiones para la validación cruzada

sfs.fit(X_train, y_train)

selected_features = list(sfs.k_feature_names_)


X_train_selected = sfs.transform(X_train)
X_test_selected = sfs.transform(X_test)

model.fit(X_train_selected, y_train)

y_pred = model.predict(X_test_selected)
mse = mean_squared_error(y_test, y_pred)
print(f"Columnas seleccionadas: {selected_features}")
print(f"Mean Squared Error: {mse}")
```
